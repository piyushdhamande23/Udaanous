```python
import pandas as pd
import numpy as np


data=pd.read_csv("/content/Messy data set.csv")
data.head()
```

```python
#sorting data
sorted_data = data.sort_values(by='Name',ascending= np.False_)
sorted_data
```

```python
#filter data

data['Age']=pd.to_numeric(data['Age'],errors ='coerce')
filter_data = data[data['Age']>25]
filter_data
```

```python
#is in
filter_data = data[data['Name'].isin(['Bob','Eve'])]
filter_data
```

```python
#loc and iloc

data.loc[0:3,'Name':'Email']
```

```python
data.iloc[0:3,0:3]
```

```python
#combining and conditions
filtered_data = data[(data['Name']=='Bob') | (data['Age']== 30)]
filtered_data
```

```python
filtered_data = data[(data['Name']=='Bob') & (data['Age']== 30)]
filtered_data
```
