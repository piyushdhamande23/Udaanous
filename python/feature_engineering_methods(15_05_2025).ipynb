```python
data = pd.DataFrame({
    'orderdate': ['10-15-2024', '01-20-2025', '05-05-2025']
})

data['orderdate'] = pd.to_datetime(data['orderdate'], format='%m-%d-%Y')
data['order_month'] = data['orderdate'].dt.month
data['order_month_name'] = data['orderdate'].dt.strftime('%B')

print(data)

```
