```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


df=pd.read_csv("/content/customers.csv")

df.head()
```

```python
df.info()
```

```python

gender_info = df['Gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_info, labels=gender_info.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
#analyze spending score by gender
x = df['Gender']
y = df['SpendingScore']

plt.bar(x,y)
plt.title('Analyze spending score by gender')
plt.xlabel('Gender')
plt.ylabel('SpendingScore')
plt.show()
```

```python
#Relationship between age and spending score
x = df['Age']
y = df['SpendingScore']

plt.bar(x,y)
plt.title("Relationship between age and score")
plt.xlabel("Age")
plt.ylabel("SpendingScore")
plt.show()
```

```python
#category wise analysis
x= df['PurchaseCategory']
y= df['SpendingScore']

plt.bar(x,y)
plt.title('Category wise analysis')
plt.xlabel("PurchaseCategory")
plt.ylabel("SpendingScore")
plt.show()
```

```python
#Income analysis
x= df['AnnualIncome']

mylabel = df['SpendingScore']

plt.pie(x,labels = mylabel)
plt.title("Income Analysis")
plt.show()
```
