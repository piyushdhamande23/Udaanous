```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/content/ecommerce_orders.csv")

df.head()
```

```python
df.dtypes
```

```python
#null and missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#Distribution of OrderValue,Discount and Revenue
plt.figure(figsize=(8,6))
sns.barplot(data=df[['OrderValue','DiscountApplied','Revenue']])
plt.title('Distribution of OrderValue,Discount and Revenue')
plt.show()
```

```python
#Revenue over time

df['Date'] = pd.to_datetime(df['OrderDate'])

df = df.sort_values('Date')

plt.figure(figsize=(8, 4))
plt.plot(df['Date'], df['Revenue'], marker='o')
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Revenue by product category

sns.barplot(x='Category',y='Revenue',data=df)
plt.title('Revenue by product category')
plt.show()
```

```python
#Analyze payment method

data = df['PaymentMethod'].value_counts()

data.plot.pie(labels = data.index ,autopct = '%1.1f%%', startangle = 90)
plt.title("Analyze payment method")
plt.show()
```

```python
#Average Revenue per order by category
df['Average_Revenue'] = df['Revenue'].mean()
# print(Average_Revenue)

sns.barplot(data=df[['Average_Revenue','OrderID','Category']])
plt.title("Average Revenue per order by category")
plt.show()
```

```python
#discount vs revenue relationship


sns.barplot(x='DiscountApplied',y='Revenue',data=df)
plt.title("Discount Vs Revenue Relationship")
plt.show()
```
