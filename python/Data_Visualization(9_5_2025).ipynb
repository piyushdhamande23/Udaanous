```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#scatterplot
data =pd.DataFrame({
'Order_ID' :[2,10,45,20,30,8,15],
'Price' : [101,52,60,100,85,56,68]
})

plt.figure(figsize=(10,6))
plt.scatter(data['Order_ID'],data['Price'])
plt.title('Order Details')
plt.xlabel('Order_ID')
plt.ylabel('Price')

plt.show()
sns.regplot(x='Order_ID',y='Price',data=data)
```

```python
#Heatmap

rand_data = np.random.rand(10, 10)
correlation_matrix = np.corrcoef(rand_data)
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True)
plt.title('matrix details')
plt.show()

```

```python
#boxplot
data =pd.DataFrame({
'Order_ID' :[2,10,45,20,30,8,15],
'Price' : [101,52,60,100,85,56,68]
})

plt.figure(figsize=(8,6))
sns.boxplot(x='Order_ID',y='Price',data=data)
plt.title('Boxplot data')
plt.xlabel('Order_ID')
plt.show()
```
