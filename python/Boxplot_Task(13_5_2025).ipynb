```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel("/content/boxplot_data.xlsx")
data.head()
```

```python
#boxplot task

sns.boxplot(x=data["Task Completion Rates"],y=data["churned"])

plt.show()

```

```python
median_df = data.groupby('churned', as_index=False)['Task Completion Rates'].median()

sns.boxplot(x='churned', y='Task Completion Rates', data=median_df)
plt.title("Boxplot of Median Values")
plt.show()
```
