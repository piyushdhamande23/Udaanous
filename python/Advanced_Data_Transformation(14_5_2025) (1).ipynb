```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import boxcox
from scipy.stats import yeojohnson


data=pd.read_excel("/content/practice_data.xlsx")
data.head()

```

```python
#boxcox transformation
from scipy.stats import shapiro,boxcox

column_name =data["Sales Revenue Generated"]

def boxcox_transformation (data,column_name):
  transformed_data, _ = boxcox(data[column_name])
  data[f'{column_name}_boxcox'] = transformed_data

  stat, p_value = shapiro(data[f'{column_name}_boxcox'])
  sns.kdeplot(data[f'{column_name}_boxcox'],fill='True')

  plt.show()

  print('P value: ', p_value)
```

```python
#yeojohnson transformation

column_name = data["Sales Revenue Generated"]
def yeojohnson_transformation(data, column_name):
 transformed_data, _ = yeojohnson(data[column_name])
 data[f'{column_name}_yeojohnson'] = transformed_data
 stat, p_value = shapiro(data[f'{column_name}_yeojohnson'])
 kdeplot = sns.kdeplot(data[f'{column_name}_yeojohnson'])

 print(kdeplot)
 print('P value: ', p_value)
```

```python
#one sample ttest
from scipy import stats


hypothesis_mean = 68

t_statistic, p_value = stats.ttest_1samp(data['Days Present'], hypothesis_mean)


print("T-statistic:", t_statistic)
print("P-value:", p_value)

alpha = 0.05


if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the average purchase amount and 68.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the average purchase amount and 68.")
```

```python
#independent sample ttest
from scipy import stats
import pandas as pd


churned_customer = data[data['churned'] == 'Yes']['Sales Revenue Generated']
existing_customer =data[data['churned'] == 'No']['Sales Revenue Generated']


t_statistic, p_value = stats.ttest_ind(churned_customer, existing_customer, equal_var=False)  # Welch's t-test


print("T-statistic:", t_statistic)
print("P-value:", p_value)


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The group means are significantly different.")
else:
    print("Fail to reject the null hypothesis. No significant difference between group means.")
```

```python
import pandas as pd
from scipy import stats


data = pd.DataFrame({
    "city": ["Chicago", "New York", "Houston", "Los_angeles"],
    "values": [4, 5, 7, 3]
})


groups = [group["values"].values for _, group in data.groupby("city")]


t_statistic, p_value = stats.f_oneway(*groups)


print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis. No significant difference between group means.")
```
