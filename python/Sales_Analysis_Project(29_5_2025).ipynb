```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/sales.csv")
df.head()
```

```python
data = df[['Region','Sales']]
plt.bar(data['Region'],data['Sales'])
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales Data by Region')
plt.show()
```

```python
data = df[['ProductName','Sales']]
plt.bar(data['ProductName'],data['Sales'])
plt.xlabel('ProductName')
plt.ylabel('Sales')
plt.title('Product Name wise Sales')
plt.show()
```

```python


plt.plot(df['OrderDate'], df['Sales'], label='Sales')
plt.plot(df['OrderDate'], df['ProductName'], label='Product Name')

plt.xlabel('OrderDate')
plt.ylabel('Sales')
plt.title('Sales Data by OrderDate')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
