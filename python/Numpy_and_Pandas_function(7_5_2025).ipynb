```python
import pandas as pd
import numpy as np

data1 = pd.DataFrame({
    'ID': [1,2,3,4,5],
    'Name': ['Alice','Bob','Willimson','David','Watson']
})

data2 = pd.DataFrame({
    'ID':[1,2,3,4,5],
    'City':['Mumbai','Pune','Nashik','Nagpur','Delhi']

})
print(data1)
print(data2)
```

```python
#Joining data
join_data = pd.merge(data1,data2,on = 'ID',how ='inner')
join_data.head()
```

```python
#concatenate data
concatenated_data = pd.concat([data1,data2])
concatenated_data.head()
```

```python
#value count
data1['Name'].value_counts()
```

```python
#group by
grouped_var = data1.groupby('Name')['ID'].mean()
grouped_var
```

```python
#pivot table
pivot_table = data1.pivot_table(index = 'Name', values = ['ID'],aggfunc='mean')

#pivot_table = pivot_table.rename(coloumn={'Name':'New_Name'})
pivot_table


```
